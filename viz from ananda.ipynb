```python
import pandas as pd
import numpy as np
import csv
import json
import plotly.graph_objects as go
import plotly.express as px
import matplotlib
```

```python
bestsellers = pd.read_csv("bestsellers with categories.csv")
bestsellers
```

```python
def truncateName(name):
    if len(name)>30:
        name = name[:30]
        name = name+". . ."
    return name

df4 = bestsellers.groupby(["Name"]).sum(["Reviews"]).reset_index()
df4 = df4.sort_values(by=['Reviews'], ascending=False)
df4 = df4.head(5)
df4 = df4[["Name", "Reviews"]]
df4["short_name"] = df4["Name"].transform(truncateName)
```

```python
df = bestsellers.groupby(["Year"])["Reviews", "User Rating"].mean().reset_index()
fig = px.scatter(df, x="Year", y="Reviews", color="User Rating",
                 title="Number of Amazon Reviews by Year with Average User Rating")
```
